```python
import numpy as np
from scipy.linalg import eigh
from qiskit.quantum_info import SparsePauliOp

N = 6
J = 1.0
h = 0.5

terminos = []

# -J * Z_i @ Z_{i+1}
for i in range(N - 1):
    terminos.append(("ZZ", [i, i + 1], -J))
terminos.append(("ZZ", [N-1, 0], -J)) # termino ciclico

# -h * X_i
for i in range(N):
    terminos.append(("X", [i], -h))

# esto lo hizo gemini
H_op = SparsePauliOp.from_sparse_list(terminos, num_qubits=N)

H = H_op.to_matrix()

val_propios, vec_propios = eigh(H)
energia_base, estado_base = val_propios[0], vec_propios[0]


print("Energía del estado base: ", energia_base)
print("Estado base: ", estado_base)
print(H_op)
```
